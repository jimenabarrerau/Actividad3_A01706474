```python
import pandas as pd
df = pd.read_excel("Gastos y costos 20-23.xlsx", sheet_name = '2023', header = 3 )
df
```

```python
df.info()
```

```python
nulos = df.isnull().sum()
nulos
```

```python
#Imputación método ffill variables categóricas
folio = df["Folio"] =df["Folio"].fillna(method="ffill")
tipo = df["TIPO GASTO"] =df["TIPO GASTO"].fillna(method="ffill")
otros = df["Otros"] =df["Otros"].fillna(method="ffill")
gasto = df["MP"] =df["MP"].fillna(method="ffill")
poliza = df["Poliza"] =df["Poliza"].fillna(method="ffill")
```

```python
#Imputación variables numéricas
tc = df["FP"]=df["FP"].fillna(round(df["TC"].mean()))
importe = df["Importe"]=df["Importe"].fillna(round(df["Importe"].mean()))
```

```python
nulos = df.isnull().sum()
nulos
```

```python
import matplotlib.pyplot as plt
fig = plt.figure(figsize = (7,7))
df.plot(kind="box", vert= False)
plt.suptitle("Outliers")
plt.show()
```

```python
superior = df.mean() + 3*df.std()
inferior = df.mean() - 3*df.std()
print("Límite superior" , superior)
print()
print("Límite inferior" , inferior)
```

```python
#Imputación valores numéricos
df2 = df[(df <= superior) & (df >= inferior)]
df2
```

```python
df3 = df2.drop(["Fecha", "Folio", "UUID", "RFC","Proveedor", "TIPO GASTO","Descripción","Tipo", "Status", "Poliza","Otros"], axis=1)
df3
```

```python
#Comprobación nulos
df3.isnull().sum()
```

```python
#sustituyo con valores de media
limpio= df3.copy()
limpio = limpio.fillna(round(df3.mean(),1))
limpio
```

```python
#Comprobación nulos
limpio.isnull().sum()
```

```python
import matplotlib.pyplot as plt
plt.boxplot(limpio)
plt.suptitle("Outliers std")
plt.show()
```

```python
per1 = limpio.quantile(0.25) #Q1
per3 = limpio.quantile(0.75) #Q3

iqr = per1 - per3

lim_sup_iqr = per3+ 1.5*iqr
lim_inf_iqr = per3- 1.5*iqr
print("Límite superior permitido" , lim_sup_iqr)
print()
print("Límite inferior permitido" , lim_inf_iqr)
```

```python
#datos limpios de df
df3_iqr = limpio[(limpio <=lim_sup_iqr)&(limpio >=lim_inf_iqr)]
df3_iqr
```

```python
df4_iqr = df3_iqr.copy()
df4_iqr = df4_iqr.fillna(round(df3_iqr.mean(),1))
df4_iqr
```

```python
#unimos columnas string eliminadas
col = ["Fecha", "Folio", "UUID", "RFC","Proveedor", "TIPO GASTO","Descripción","Tipo", "Status", "Poliza","Otros"]
indice= df[col]
limpios = pd.concat([indice, df4_iqr],axis=1)
limpios
```

```python
import matplotlib.pyplot as plt
plt.boxplot(df4_iqr)
plt.suptitle("Outliers intercuartil")
plt.show()
```
